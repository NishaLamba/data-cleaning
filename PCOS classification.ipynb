```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Checking for wrong entries like symbols -,?,#,*,etc.
for col in df.columns:
    print('{} : {}'.format(col,df[col].unique()))
```

```python
for col in df.columns:
    df[col].replace({'?':np.nan},inplace=True)
```

```python
df.isnull().sum()
```

# Visualizing the missing values

```python
plt.figure(figsize=(30,10))

sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Blues')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x='Pimples(Y/N)',y='PCOS (Y/N)',data=df)
plt.xlabel('Pimples(Y/N)')
plt.ylabel('PCOS (Y/N)')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x='Pulse rate(bpm)',y='PCOS (Y/N)',data=df)
plt.xlabel('Higway mpg')
plt.ylabel('Price')
```

```python
plt.figure(figsize=(30,10))
sns.boxplot(x='PCOS (Y/N)',y='Age (yrs)',data=df)
```

```python
# visualize the target variable
g = sns.countplot(df['PCOS (Y/N)'])
g.set_xticklabels(['1','0'])
plt.show()
```
